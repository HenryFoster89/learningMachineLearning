# Supervised ML
- Classification

Predict a class label, classification could be Binary or Multilclass

- Regression

Predict a continous number

In supervised we want to build a model on the training data and then be able to make accurate predictions on new data.
- Generalitation: a model able to make accurate predictions on new data
- Overfitting: when you fit a model too closely to the particularities of the training set and obtain a model that works well on the training set but is not able to generalize to new data
- Underfitting: if your model is too simple when you might not be able to capture all the aspects of and variability in the data, and your model will do badly even on the training set.

## K-Nearest Neighbors
### Introduction

KNN is a typical example of a lazy learner. It is called "lazy" not because of its apparent simplicity, but because it doesn't learn a discriminative function from the training data but memorizes the training dataset instead.
KNN belongs to a subcategory of nonparametric models that is described as instance-based learning. Models based on instance-based learning are characterized by memorizing the training dataset, and lazy learning is a special case of instancebased learning that is associated with no (zero) cost during the learning process.

The KNN algorithm itself is fairly straightforward and can be summarized by the
following steps:
1. Choose the number of k and a distance metric.
2. Find the k-nearest neighbors of the data record that we want to classify.
3. Assign the class label by majority vote.

Based on the chosen distance metric, the KNN algorithm finds the k examples in the training dataset that are closest (most similar) to the point that we want to classify.
The class label of the data point is then determined by a majority vote among its k nearest neighbors. The main advantage of such a memory-based approach is that the classifier immediately adapts as we collect new training data. However, the downside is that the computational complexity for classifying new examples grows linearly with the number of examples in the training dataset in the worst-case scenariounless the dataset has very few dimensions (features) and the algorithm has been implemented using efficient data structures such as k-d trees.

### Applying in Scikit-Learn
#### Low dimensional model

```python
# Generate dataset
X, y = mglearn.datasets.make_forge()

# Dividi in train-test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Instanziare la classe con i parametri 
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)

# Fit il classifier
clf.fit(X_train, y_train)

# Make prediction
clf.predict(X_test)

# Verify the prediction
print(f"Test set accuracy: {clf.score(X_test, y_test):.2f}")
```

#### K parameter

```python
fig, axes = plt.subplots(1,3, figsize = (12,3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # Initialize class
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)

    # Plot
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)

    # Chart format
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")

axes[0].legend(loc=3)
```

Figura 2-1: Complessit del Modello e Generalizzazione

Il primo grafico rappresenta il compromesso tra la **complessit del modello** e l'**accuratezza della generalizzazione**. (Il grafico non c'  a pagina 29 del libro)

Cosa mostra il grafico?
- **Modelli troppo semplici (bassa complessit)**
  - Soffrono di **underfitting**.
  - Scarsa accuratezza sia sui dati di **addestramento** che su quelli di **test**.
- **Modelli troppo complessi (alta complessit)**
  - Soffrono di **overfitting**.
  - Alta accuratezza sui dati di **training**, ma scarsa capacit di **generalizzazione**.
- **Il "Sweet Spot"**
  -  il livello di complessit **ottimale**, in cui il modello generalizza meglio.

Figura 2-6: Effetto della Scelta di `k` in KNN

Questo grafico mostra come il valore di `k` influisce sulla decision boundary in un modello **K-Nearest Neighbors (KNN)**.

| Grafico | Numero di Vicini (`k`) | Effetto |
|---------|------------------|---------|
| **Sinistra** | `k = 1` | Il modello segue troppo da vicino i dati di training  **Overfitting**. La decision boundary  molto frastagliata. |
| **Centro** | `k = 3` | Decision boundary pi regolare, buon compromesso tra flessibilit e generalizzazione. |
| **Destra** | `k = 9` | Decision boundary molto smussata, tende a generalizzare troppo  **Underfitting**. |

Conclusione

- Un valore **basso** di `k` rende il modello pi **complesso** e incline all'**overfitting**.
- Un valore **alto** di `k` rende il modello pi **semplice**, ma pu portare a **underfitting**.
- La scelta ottimale di `k` dipende dai dati e deve essere valutata con metodi come la **validazione incrociata**.

#### Real world model 

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

# try n_neighbors from 1 to 10
neighbors_settings = range(1, 25)
for n_neighbors in neighbors_settings:

    # build the model
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)

    # record training set accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    
    # record generalization accuracy
    test_accuracy.append(clf.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.show()
plt.close()
```

Interpretazione del Grafico

- **Training Accuracy (linea blu)**: 
  - Con `k = 1`, l'accuratezza del training  **perfetta** (100%), segno di **overfitting**.
  - Aumentando `k`, il modello diventa pi semplice e l'accuratezza del training diminuisce.
- **Test Accuracy (linea arancione)**: 
  - Con `k = 1`, l'accuratezza del test  pi bassa rispetto ad altri valori di `k`, segno che il modello  troppo complesso e non generalizza bene.
  - Con `k` intorno a **6**, l'accuratezza sul test  massima: questo rappresenta un **buon compromesso** tra complessit e generalizzazione.
  - Con `k = 10`, l'accuratezza del test peggiora, indicando **underfitting**.

Conclusione

- `k` piccolo  Modello troppo complesso (**overfitting**).
- `k` grande  Modello troppo semplice (**underfitting**).
- Il valore ottimale di `k`  intorno a **6** nel dataset analizzato.

### k-neighbors regression 

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = mglearn.datasets.make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# instantiate the model and set the number of neighbors to consider to 3
reg = KNeighborsRegressor(n_neighbors=3)

# fit the model using the training data and training targets
reg.fit(X_train, y_train)

print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(X_train, y_train, '^', c="blue", markersize=8)
ax.plot(X_test, y_test, 'v', c="red", markersize=8)
    
ax.legend(["Training data/target", "Test data/target"], loc="best")

plt.show()
plt.close()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 14))

# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

for n_neighbors, ax in zip([1, 3, 9], axes):

    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)

    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)

    ax.set_title(
    "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
    n_neighbors, reg.score(X_train, y_train),
    reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
    
axes[0].legend(["Model predictions", "Training data/target",
    "Test data/target"], loc="best")

plt.tight_layout()
plt.show()
plt.close()
```

## Linear model


I modelli lineari sono una classe di modelli ampiamente utilizzati nella pratica e studiati da decenni, con radici che risalgono a oltre cento anni fa.

### Caratteristiche principali
- Fanno previsioni utilizzando una **funzione lineare** delle feature di input.
- Sono semplici da interpretare e implementare.
- Funzionano bene su dati in cui la relazione tra feature e target  **approssimativamente lineare**.

### Vantaggi
- **Efficienza computazionale**
  - Possono essere addestrati rapidamente anche su grandi dataset.
- **Facilit di interpretazione**
  - Coefficienti del modello indicano il contributo di ciascuna feature alla previsione.
- **Generalizzazione**
  - Con una buona regolarizzazione, possono generalizzare bene su dati non visti.

### Svantaggi
- **Limitati alle relazioni lineari**
  - Se la relazione tra feature e target  non lineare, il modello potrebbe non essere adeguato.
- **Sensibili agli outlier**
  - Gli outlier possono avere un impatto significativo sui coefficienti del modello.

I modelli lineari sono una base solida per molte tecniche di machine learning e spesso vengono utilizzati come punto di partenza prima di considerare modelli pi complessi.

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(X, y)

plt.xlabel("X")
plt.ylabel("y")

plt.show()
plt.close()
```

```python
X_b = np.c_[np.ones((100, 1)), X] # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)


X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new] # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(X_new, y_predict, "r-")

plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# Valori stimati per intercetta e coefficiente
intercept = lin_reg.intercept_[0]
coef = lin_reg.coef_[0, 0]
print(f"Intercetta: {intercept:.2f}, Coefficiente: {coef:.2f}")

# Creazione della linea di regressione
X_new = np.linspace(0, 2, 100).reshape(-1, 1)
y_pred = lin_reg.predict(X_new)

# Plot dei dati e della retta di regressione
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(X, y, color='blue', alpha=0.6, label="Dati")
ax.plot(X_new, y_pred, color='red', linewidth=2, label="Regressione Lineare")

ax.set_xlabel("X")
ax.set_ylabel("y")
ax.legend()
plt.show()
```
